# 平面图像的超分辨率方法

### SRCNN
Image Super-Resolution Using Deep Convolutional Networks, ECCV 2014

首次将深度学习应用于超分辨率重建，提出基于卷积神经网络的超分辨率算法。使用三层卷积神经网络就在超分辨率重建质量和效率等方面超越了大部分基于示例的超分辨率算法，成为当时最优的算法。

## 基于网络结构优化的超分辨率算法

基于深度学习的超分辨率算法通过优化网络结构，能够有效提升超分辨率重建质量。研究人员从加深网络层次、增加稠密连接、模块化残差结构等多个维度对超分辨率网络进行优化。

### VDSR
Accurate Image Super-Resolution Using Very Deep Convolutional Networks, CVPR 2016

第一个将残差学习的思想引入超分辨率重建的方法，构建了深度卷积超分辨率网络。采用残差学习，将网络结构加深到20层，有效提升超分辨率重建的质量和效率。


### EDSR
Enhanced Deep Residual Networks for Single Image Super-Resolution， CVPR 2017

提出增强深度残差超分辨率网络。通过删除超分辨率网络中批规范化处理BN(Batch Normalization)降低超分辨率模型对计算资源的消耗，从而将网络深度增加到80层，有效提升了超分辨率重建的质量。

## 基于生成对抗的超分辨率算法

生成对抗网络(Generative Adversarial Network, GAN)通过生成器和鉴别器的对抗训练，能够生成视觉感知自然、清晰的图像。



## SRGAN
Photo-Realistic Single Image Super-Resolution Using a Generative Adversarial Network, CVPR 2017

首次将生成对抗网络用于超分辨率问题。


## 基于注意力机制的超分辨率算法

注意力机制能够使超分辨率网络专注于纹理细节信息丰富的通道，提升对图像上下文信息的利用能力，能够哦有效提升超分辨率重建的质量。


## RCAN

Image Super-Resolution Using Very Deep Residual Channel Attention Networks, ECCV 2018

提出一种基于一阶注意力网络的超分辨率算法，采用通道注意力机制，使得超分辨率网络专注于更具有丰富特征的通道。通过赋予高频信息丰富的通道较大的权重，提升卷积神经网络对高频信息的感知能力和图像上下文信息的利用能力。

将 Swin Transformer 应用到图像恢复工作中，在降低参数量的同时取得很好的效果。


# 全景图像超分辨率算法

通常全景图像的原始图像格式(如3D球面)必须被转换成2D平面图像表示以进行存储和传输。等距柱状投影(Equirectangular projection ERP)是最常用的投影格式，原球面图像的经纬度被映射到横纵坐标网格上。ERP的采样密度是不均匀的，因而随着纬度产生失真。全景图像的采样与失真问题由原始球面图像与2D平面图像之间的转换引起。

## 基于球面-平面映射先验信息的超分辨率算法

### 360-SS

Super-resolution of Omnidirectional Images Using Adversarial Learning, MMSP 2019

提出了基于球面加权的损失函数来应对ERP图像的失真问题。

### OSRT

Omnidirectional Image Super-Resolution with Distortion-aware Transformer, CVPR 2023

根据球面失真图计算偏移量，通过可变形卷积使网络自适应地选取采样点以解决ERP图像的失真问题。

### LAU-Net

Latitude Adaptive Upscaling Network for Omnidirectional Image Super-resolution, CVPR 2021
